###### Hàm in danh sách giải

Sử dụng **vòng lặp for** để hiển thị toàn bộ danh sách giải.

```python
def in_danh_sach_giai(danh_sach_giai):
    # Truyền biến cục bộ giải để hiển thị số giải
    giai = 1
    for so in danh_sach_giai:
        print(f"Giai {giai}: {so}")
        giai += 1

# Truyền kết quả của hàm danh sách giải vào biến danh_sach_giai
danh_sach_giai = quay_so()
# Chạy hàm in_danh_sach_giai
in_danh_sach_giai(danh_sach_giai)
```

Trong string có hàm **isdigit()** để kiểm tra có phải số hay không. Ta sẽ kết hợp để hoàn thành hàm nhập số lô đầy đủ:

```python
def nhap_so_lo():
    while True:
        so_lo = input("Nhập số lô mà bạn muốn cược (10-99): ")
        if so_lo.isdigit() and 10 <= int(so_lo) <= 99:
            return int(so_lo)
        else:
            print("Nhập sai định dạng, vui lòng nhập lại!")

nhap_so_lo()
```

###### Hàm nhập tiền cược
Ta cần khai báo biến toàn cục **tong_tien** để số tiền sẽ giữ giá trị  khi đi qua một hàm nào đó và có thể sử dụng biến toàn cục tại một hàm khác mà không phải khai báo lại.

```python
# VD1: Không dùng biến toàn cục
def tinh_tong_tien():
  tong_tien = 1000000
  tien_trung_giai = 1000000
  tong_tien = tong_tien + tien_trung_giai
  print("Tổng tiền trong hàm = ", tong_tien)


tinh_tong_tien()

# In tổng tiền ngoài hàm sẽ bị lỗi
print(tong_tien)
```

```python
# VD2: Sử dụng biến toàn cục
tong_tien = 1000000
def tinh_tong_tien():
  # Không cần gán lại giá trị tong_tien nữa, chỉ cần báo đây là biến toàn cục
  global tong_tien

  tien_trung_giai = 1000000
  tong_tien = tong_tien + tien_trung_giai
  print("Tổng tiền trong hàm = ", tong_tien)


tinh_tong_tien()
# In tổng tiền ngoài hàm xem có khác gì không
print("Tổng tiền ngoài hàm = ", tong_tien)
```

Áp dụng kiến thức trên, ta xây dựng được hàm **nhap_tien_cuoc**

```python
def nhap_tien_cuoc():
    while True:
        tien_cuoc = input("Nhập tiền cược mà bạn muốn cược (tiền cược < tổng tiền bạn có): ")
        if tien_cuoc.isdigit() and int(tien_cuoc) < tong_tien:
            return int(tien_cuoc)
        else:
            print("Nhập sai định dạng hoặc quá tổng tiền. Nhập lại")
```

###### Hàm tổng hợp và xử lý

Để biết trúng lô hay không, ta chỉ cần so sánh số lô người chơi đã cược và các con số trong danh sách giải. Ta sử dụng thêm một biến cục bộ dem_so_lo để đếm tổng số lô trúng.

Nếu trúng lô, tổng tiền sẽ bằng tiền cược * số lô trúng * 70
Nếu "tạch lô", tổng tiền sẽ trừ đi số tiền đã cược

```python
def game_quay_so(so_lo, tien_cuoc):
    global tong_tien

    danh_sach_giai = quay_so()
    dem_so_lo = 0
    print(f"Số lô bạn chơi là: {so_lo}")
    in_danh_sach_giai(danh_sach_giai)
    for giai in danh_sach_giai:
        if so_lo == giai:
            dem_so_lo += 1
    if dem_so_lo > 0:
        tong_tien += (tien_cuoc * dem_so_lo) * 70
        print(f"Bạn đã trúng {dem_so_lo} nháy!")
        print(f"Số tiền bạn trúng là {(tien_cuoc * dem_so_lo) * 70}")
        print(f"Tổng tiền bạn có là {tong_tien}")
    else:
        print("Bạn đã thua lô!")
        tong_tien -= tien_cuoc
        print(f"Tổng tiền bạn còn là {tong_tien}")

```

##### Chức năng nạp tiền

Ta áp dụng kiến thức đã giới thiệu ở trên về **biến toàn cục**, hàm **while - break - contine** để cho phép người dùng nạp tiền và hiển thị tổng tiền

```python
def nap_tien():
    global tong_tien
    while True:
        tien_nap = int(input("Nhập số tiền mà bạn muốn nạp: "))
        tong_tien += tien_nap
        print(f"Tổng tiền sau khi bạn nạp là {tong_tien}")
        nap_tiep = int(input("Bạn có muốn nạp tiếp không? (0: không, 1: có)"))
        if nap_tiep == 0:
            break
        else:
            continue
```

#### Phía hiển thị cho người dùng

Ta xây dựng hàm menu() để hiện thị menu cho người dùng lựa chọn

```python
def menu():
    print("GAME LÔ ĐỀ HỌC")
    print("1. Chơi lô")
    print("2. Nạp tiền")
    print("3. Thoát game")
```

Xây dựng hàm main, cho phép người dùng nhập vào lựa chọn sử dụng **while** và **match case**. Ngoài ra, ta sử dụng **break** để thoát khỏi vòng lặp khi người dùng đã thực hiện lựa chọn xong.

**Lưu ý** hàm **match case** chỉ có trên Python 3.10 trở lên

```python
def main():
    menu()
    while True:
        lua_chon = int(input("Bạn chọn: "))
        match lua_chon:
            case 1:
                print("Bạn chọn lựa chọn 1")
                break
            case 2:
                print("Bạn chọn lựa chọn 2")
                break
            case 3:
                print("Bạn chọn lựa chọn 3")
                break
            case _:
                print("Nhập sai lựa chọn, vui lòng nhập lại !")
```

```python
main()
```

Ghép các hàm vào hàm **match case** và hoàn thiện hàm main

```python
def main():
    menu()
    while True:
        lua_chon = int(input("Bạn chọn: "))
        match lua_chon:
            case 1:
                clear_output()
                while True:
                    tien_cuoc = nhap_tien_cuoc()
                    so_lo = nhap_so_lo()
                    game_quay_so(so_lo, tien_cuoc)
                    choi_tiep = int(input("Bạn có muốn chơi tiếp không? (0: không, 1: có)"))
                    if choi_tiep == 0:
                        break
                clear_output(wait=True)
                main()
                break
            case 2:
                clear_output()
                nap_tien()
                clear_output(wait=True)
                main()
                break
            case 3:
                print("Chào bạn, hẹn gặp lại!")
                break
            case _:
                print("Nhập sai lựa chọn, vui lòng nhập lại !")
```
