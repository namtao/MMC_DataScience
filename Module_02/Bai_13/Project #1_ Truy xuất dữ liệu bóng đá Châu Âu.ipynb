# Gọi thư viện Python

```python
#Import libraries
import numpy as np
import pandas as pd
import sqlite3
from datetime import timedelta
import warnings
warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", 100)
```

## Câu hỏi 1: Kết nối tới cơ sở dữ liệu và truy vấn thông tin của tất cả các bảng

Đọc https://docs.python.org/2/library/sqlite3.html

Học viên chỉ sử dụng câu lệnh SQL để truy vấn kế quả, không được sử dụng câu lệnh Pandas để tính toán ra kết quả.

```python
# Kết nối với cơ sở dữ liệu
conn = sqlite3.connect('./Data/database.sqlite')

# Lấy thông tin của tất cả bảng
tables = pd.read_sql("""SELECT *
                        FROM sqlite_master
                        WHERE type='table';""", conn)
tables
```

## Câu hỏi 2: Liệt kê các quốc gia có trong dữ liệu bảng "Country"

```python
result = pd.read_sql("select * from country", conn)
result
```

## Câu hỏi 3: Liệt kê các giải đấu trong bảng "League"

```python
result = pd.read_sql("select * from league", conn)
result
```

## Câu hỏi 4: Liệt kê các giải đấu và thông tin quốc gia của từng league

Bảng League có quan hệ với bảng Country qua khóa country_id.

```python
result = pd.read_sql(
    "select * from league l join country c on l.country_id = c.id", conn
)
result
```

## Câu hỏi 5: Liệt kê các trận đấu trong bảng  “Match”
- Khi một đội là chủ nhà, đội đó là "home team". Đội còn lại là "away team"
- Trong bảng Match, mỗi dòng dư liệu là một trận đấu của một đội chủ nhà và đội khách, tương ứng với kết quả ghi bàn đội nhà (home goal) và đội khách (away goal).

```python
result = pd.read_sql("select * from match", conn)
result
```

## Câu hỏi 6: Liệt kê các trận đấu và thông tin về leage và country tương ứng

Bảng Match có quan hệ với :
 - Bảng Country thông qua khóa country_id
 - Bảng League thông qua khóa League_id

Sử dụng câu lệnh join để kết nối dư liệu từ 3 bảng trên.


```python
result = pd.read_sql(
    "select * from match m join country c on m.country_id = c.id join league l on m.league_id = l.id",
    conn,
)
result
```

## Câu hỏi 7: Liệt kê số lượng trận đấu của mỗi giải đấu bao gồm cả tên giải đấu, sắp xếp theo thứ tự giảm dần số trận đấu

```python
result = pd.read_sql(
    "select name, count(name) as 'SL' from match m join league l on m.league_id = l.id group by name order by name",
    conn,
)
result
```

## Câu hỏi 8: Liệt kê tổng số bàn thắng của đội nhà và đội khách trong mỗi giải đấu
- Sử dụng câu lệnh Group By

```python
result = pd.read_sql(
    "select league_id, sum(home_team_goal), sum(away_team_goal) from match group by league_id, home_team_goal, away_team_goal",
    conn,
)
result
```

## Câu hỏi 9: Liệt kê thông tin các đội từ bảng “Team”

```python
result = pd.read_sql(
    "select * from team",
    conn,
)
result
```

## Câu hỏi 10: Liệt kê 20 đội với số bàn thắng sân nhà cao nhất
- Sử dung câu lênh GROUP BY, LIMIT

```python
result = pd.read_sql(
    "select team_long_name, sum(home_team_goal) from match m join team t on m.home_team_api_id = t.team_api_id group by team_long_name, home_team_goal order by sum(home_team_goal) desc limit 20",
    conn,
)
result
```

## Câu hỏi 11: Liệt kê 20 đội có số bàn thắng trên sân khách cao nhất
- Sử dụng câu lênh GROUP BY, LIMIT

```python
result = pd.read_sql(
    "select team_long_name, sum(away_team_goal) from match m join team t on m.away_team_api_id = t.team_api_id group by team_long_name, away_team_goal order by sum(away_team_goal) desc limit 20",
    conn,
)
result
```

## Câu hỏi 12: Liệt kê tên các đội bóng và tổng số bàn thắng mỗi đội, sắp xếp  theo số lượng giảm dần

- Tổng số bàn thắng của một đội bao gồm cả số bàn thắng ở sân nhà và sân khách.
- Sử dụng câu lệnh Union  

```python
result = pd.read_sql(
    """select a.ten, sum(a.slbt) as bt from (select team_long_name as ten, sum(away_team_goal) as slbt from match m join team t on m.away_team_api_id = t.team_api_id group by team_long_name, away_team_goal  
    union 
    select team_long_name, sum(home_team_goal) from match m join team t on m.home_team_api_id = t.team_api_id group by team_long_name, home_team_goal) a group by a.ten""",
    conn,
)
result
```

## Câu hỏi 13: Liệt kê tên các đội bóng (long name) và tổng số trận đấu đội đó tham gia, sắp xếp theo thứ tự giảm dần.

```python
result = pd.read_sql(
    "select team_long_name, count(*) as sl from match m join team t on m.home_team_api_id = t.team_api_id group by team_long_name order by team_long_name",
    conn,
)
result
```

## Câu hỏi 14: Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"


```python
result = pd.read_sql(
    """select a.team_long_name, a.info, count(*) as 'sl' from (select team_long_name, CASE
    WHEN home_team_goal > away_team_goal THEN 'Thang'
    WHEN home_team_goal < away_team_goal THEN 'Thua'
    ELSE 'Hoa'
    END AS info 
    
    from  match m join team t on m.home_team_api_id = t.team_api_id) a
    group by a.team_long_name, a.info
    """,
    conn,
)
result
```

## Câu hỏi 15: Liệt kê 10 đội có số trận thắng nhiều nhất

```python
result = pd.read_sql(
    """select a.team_long_name, a.info, count(*) as 'sl' from (select team_long_name, CASE
    WHEN home_team_goal > away_team_goal THEN 'Thang'
    WHEN home_team_goal < away_team_goal THEN 'Thua'
    ELSE 'Hoa'
    END AS info 
    
    from  match m join team t on m.home_team_api_id = t.team_api_id) a
    where a.info = 'Thang'
    group by a.team_long_name, a.info
    order by count(*) desc
    limit 10
    """,
    conn,
)
result
```

## Câu hỏi 16: Liệt kê tỉ lệ phần trăm về số trận của mỗi giải trên tổng tất cả các trận

```python
result = pd.read_sql(
    """
    SELECT
        name,
        (COUNT(name) * 100.0 / (SELECT COUNT(*) FROM match)) AS percentage
    FROM 
        match m join league l on m.league_id = l.id
    GROUP BY 
        name
    order by percentage
    """,
    conn,
)


result
```

## Câu hỏi 17: Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận.

```python
result = pd.read_sql(
    """SELECT
        name, (sum(home_team_goal) + sum(away_team_goal))*100.0/ (select sum(home_team_goal) + sum(away_team_goal) from match) as percentage
    FROM 
        match m join league l on m.league_id = l.id
    group by name
    order by percentage
    """,
    conn,
)
result
```

## Câu hỏi 18: Liệt kê tổng số bàn ghi được của từng giải đấu theo từng mùa giải.

```python
result = pd.read_sql(
    """SELECT
        season, sum(home_team_goal) + sum(away_team_goal) as sl
    FROM 
        match m join league l on m.league_id = l.id
    group by season
    """,
    conn,
)
result
```

## Câu hỏi 19: Liệt kê thông tin cầu thủ:

- Quy đổi cân nặng sang kilogram
- Quy đổi chiều cao sang mét
- Tính chỉ số bmi = ( (weight* 0.453592) / (height/100)^2)
- Tính tuổi của cầu thủ

```python
result = pd.read_sql(
    """
    SELECT *, (weight * 0.453592)/(height/100) as bmi, CAST((strftime('%Y', 'now') - strftime('%Y', birthday)) - (strftime('%m-%d', 'now') < strftime('%m-%d', birthday)) AS INT) AS age
    from player
    """,
    conn,
)
result
```

## Câu hỏi 20: Liệt kê cầu thủ lớn tuổi nhất

```python
result = pd.read_sql(
    """
    SELECT *, (weight * 0.453592)/(height/100) as bmi, CAST((strftime('%Y', 'now') - strftime('%Y', birthday)) - (strftime('%m-%d', 'now') < strftime('%m-%d', birthday)) AS INT) AS age
    from player
    order by age desc
    limit 1
    """,
    conn,
)
result
```
