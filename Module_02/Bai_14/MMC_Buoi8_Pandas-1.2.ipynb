## Đọc dữ liệu

Pandas hỗ trợ đọc dữ liệu từ nhiều nguồn khác nhau, bao gồm:
* File CSV: `pd.read_csv(filename)`
* File TSV: `pd.read_table(filename)`
* File Excel: `pd.read_excel(filename)`
* Cơ sở dữ liệu SQL: `pd.read_sql(query, connection_object)`
* JSON: `pd.read_json(json_string)`
* HTML: `pd.read_html(url)`
* Clipboard: `pd.read_clipboard()` (dữ liệu trong clipboard sẽ được chuyển tới  `read_table())`
* Từ điển Python: `pd.DataFrame(dict)`

Ví dụ: Đọc dữ liệu từ file CSV 'california_housing_test.csv'

## Xuất dữ liệu

Pandas cũng cho phép xuất dữ liệu từ DataFrame sang các định dạng khác nhau:
* File CSV: `df.to_csv(filename)`
* File Excel: `df.to_excel(filename)`
* SQL: `df.to_sql(table_name, connection_object)`
* JSON: `df.to_json(filename)`

```python
# Xuất dữ liệu DataFrame sang file CSV
df.to_csv('california_housing_test.csv')

# Xuất dữ liệu DataFrame sang file Excel
df.to_excel('california_housing_test.xlsx')

# Xuất dữ liệu DataFrame sang SQL
from sqlalchemy import create_engine

engine = create_engine('sqlite:///california_housing_test.db')
df.to_sql('housing', engine)

# Xuất dữ liệu DataFrame sang JSON
df.to_json('california_housing_test.json')
```

## Xem và kiểm tra dữ liệu

Pandas cung cấp nhiều hàm để xem và kiểm tra dữ liệu:
* Xem n dòng đầu tiên: `df.head(n)`
* Xem n dòng cuối cùng: `df.tail(n)`
* Lấy kích thước DataFrame (số dòng, số cột): `df.shape`
* Xem thông tin về index, kiểu dữ liệu, dung lượng: `df.info()`
* Tổng kết thống kê cho các cột số: `df.describe()`
* Đếm giá trị duy nhất (cho Series): `s.value_counts(dropna=False)`
* Đếm giá trị duy nhất cho tất cả các cột: `df.apply(pd.Series.value_counts)`

```python
# Đọc file CSV
df = pd.read_csv('/content/california_housing_test.csv')
```

```python
# Xem 3 dòng đầu tiên
print(df.head(3))
```

```python
# Xem 3 dòng cuối cùng
print(df.tail(3))
```

```python
# Lấy kích thước DataFrame
print(df.shape)
```

```python
# Xem thông tin về index, kiểu dữ liệu, dung lượng
print(df.info())
```

```python
# Tổng kết thống kê cho các cột số
print(df.describe())
```

```python
# Đếm giá trị duy nhất cho tất cả các cột
for col in df.columns:
  print(df[col].value_counts(dropna=False))
```

```python
df.apply(pd.Series.value_counts)
```

## Trích xuất dữ liệu

Bạn có thể trích xuất dữ liệu từ DataFrame bằng nhiều cách:
* Lấy một cột (trả về Series): `df[col]`
* Lấy nhiều cột (trả về DataFrame): `df[[Col1, Col2]]`
* Chọn dữ liệu theo vị trí: `s.iloc[0]`
* Chọn dữ liệu theo index: `s.loc['index_one']`
* Chọn dữ liệu ở dòng đầu tiên: `df.iloc[0, :]`
* Chọn dữ liệu ở dòng đầu tiên, ô thứ nhất: `df.iloc[0, 0]`

```python
# Đọc file CSV
df = pd.read_csv('/content/sample_data/california_housing_test.csv')

# Lấy một cột (trả về Series): df[col]
col = 'longitude'
series = df[col]
print(series.head())
```

```python

# Lấy nhiều cột (trả về DataFrame): df[[Col1, Col2]]
cols = ['longitude', 'latitude']
df_subset = df[cols]
print(df_subset.head())
```

```python

# Chọn dữ liệu theo vị trí: s.iloc[0]
first_element = series.iloc[0]
print(first_element)
```

```python
# Chọn dữ liệu theo index: s.loc['index_one']
first_index = series.index[0]
first_element_by_index = series.loc[first_index]
print(first_element_by_index)
```

```python
# Chọn dữ liệu ở dòng đầu tiên: df.iloc[0, :]
first_row = df.iloc[0, :]
print(first_row)
```

```python
# Chọn dữ liệu ở dòng đầu tiên, ô thứ nhất: df.iloc[2, 4]
first_element_first_row = df.iloc[2, 4]
print(first_element_first_row)
```
