# Hướng dẫn sử dụng Lambda trong Python


Lambda là một hàm ẩn danh (anonymous function) trong Python, được sử dụng để tạo các hàm nhỏ gọn mà không cần định nghĩa một hàm riêng biệt.



## Cú pháp:
```
lambda arguments: expression
```


## Ví dụ:




### 1. Hàm lambda đơn giản:

````python
square = lambda x: x * x
print(square(5))  # Output: 25
````

### 2. Sử dụng lambda với map():


````python
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x * x, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]|
````


### 3. Sử dụng lambda với filter():


````python
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4]
````

### 4. Lambda với nhiều tham số:


````python
add = lambda x, y: x + y
print(add(10, 20))  # Output: 30
````


## Ưu điểm của lambda:


* Ngắn gọn: Tạo hàm nhanh chóng và dễ dàng mà không cần định nghĩa hàm riêng biệt.
* Linh hoạt: Có thể được sử dụng như một đối số cho các hàm khác như map(), filter(), reduce().


## Lưu ý:


* Hàm lambda chỉ có thể chứa một biểu thức.
* Hàm lambda không thể chứa các câu lệnh phức tạp như vòng lặp hay câu lệnh điều kiện.



## Kết luận:


* Lambda là một công cụ mạnh mẽ trong Python, giúp bạn viết code ngắn gọn và hiệu quả hơn.
* Nắm vững cú pháp và cách sử dụng lambda sẽ giúp bạn linh hoạt hơn trong việc xử lý dữ liệu và tạo các hàm tùy chỉn
