# Problem 1: Tạo hai ma trận 3x3 với các phần tử integer ngẫu nhiên từ 1 - 10.

```python
arrayA = np.random.randint(1, 10, (3, 3))
arrayB = np.random.randint(1, 10, (3, 3))

print("3x3 Array:")
print(arrayA)
print(arrayB)
```

# Problem 2: Thực hiện phép cộng phần tử, nhân từng phần tử, tích vô hướng với hai ma trận vừa được tạo.


```python
# Your code here
summ = arrayA + arrayB
print("Tổng từng phần tử:")
print(summ)
```

```python
multi = arrayA * arrayB
print("Tích từng phần tử:")
print(multi)
```

```python
# matmul = arrayA @ arrayB
matmul = np.dot(arrayA, arrayB)
print("Tích vô hướng:")
print(matmul)
```

# Problem 3: VIết hàm nhận vào một mảng Numpy, sau đó trả về tổng tất cả các phần tử của chúng


```python
# Your code here
def array_sum(arr):
    return np.sum(arr)

test_array = np.array([[1, 2, 3], [4, 5, 6]])
print("Tổng tất cả các phần tử trong mảng: ", array_sum(test_array))
```
