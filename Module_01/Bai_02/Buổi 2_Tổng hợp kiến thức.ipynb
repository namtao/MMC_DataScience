### 2.1 Biến Toàn Cục (Global Variables)
- Biến toàn cục là biến được khai báo ở phạm vi cao nhất trong chương trình.
- Có thể truy cập từ bất kỳ đâu trong chương trình.

````python
global_var = 10  # Đây là biến toàn cục

def function():
    print(global_var)  # Biến toàn cục có thể được truy cập trong hàm

function()
````

### 2.2 Biến Cục Bộ (Local Variables)
- Biến cục bộ là biến được khai báo bên trong một hàm hoặc một khối mã cụ thể.
- Chỉ có thể truy cập được từ bên trong hàm hoặc khối mã đó.

````python
def function():
    local_var = 20  # Đây là biến cục bộ
    print(local_var)

function()

# Lỗi sẽ xảy ra nếu cố gắng truy cập biến cục bộ từ bên ngoài hàm
print(local_var)
````

### 6.5 **Cấu trúc Match-Case** trong Python cho phép so khớp nhiều mẫu và thực thi các khối mã khác nhau dựa trên mẫu nào phù hợp.

match biểu_thức:
   case mẫu_1:
      # Khối mã thực thi nếu mẫu_1 phù hợp
   case mẫu_2:
      # Khối mã thực thi nếu mẫu_2 phù hợp
   case mẫu_3:
      # Khối mã thực thi nếu mẫu_3 phù hợp
   ...
   case _:
      # Khối mã thực thi nếu không có mẫu nào phù hợp




````python
def describe_number(x):
    '''
    hàm `describe_number` so khớp giá trị của `x` với các mẫu khác nhau và thực thi
    các khối mã tương ứng. Nếu không có mẫu nào phù hợp, mẫu `_` sẽ được thực thi.
    '''
    match x:
        case 0:
            return "Zero"
        case 1:
            return "One"
        case 2:
            return "Two"
        case _:
            return "Other"

print(describe_number(0))  # Output: Zero
print(describe_number(1))  # Output: One
print(describe_number(5))  # Output: Other
````

##7. Vòng lặp và từ khóa

###7.1. Vòng lặp `for` và `while`:


   - `for`: Dùng để lặp qua một dãy giá trị cụ thể.

````python
for i in range(5):
    print(i)
````

   - `while`: Dùng để lặp cho đến khi một điều kiện cụ thể không còn đúng nữa.


````python
i = 0
while i < 5:
   print(i)
   i += 1
````

###7.2. Từ khóa `break` và `continue`:
   - `break`: Dùng để thoát khỏi vòng lặp ngay lập tức khi một điều kiện nhất định được đáp ứng.


````python
for i in range(10):
    if i == 5:
        break
    print(i)
````

   - `continue`: Dùng để bỏ qua phần còn lại của vòng lặp và chuyển đến lần lặp tiếp theo mà không thực hiện các câu lệnh trong khối mã dưới nó.

````python
for i in range(10):
    if i == 5:
        continue
    print(i)
````

###7.3. Từ khóa `assert`:
   - Dùng để kiểm tra điều kiện nếu là True thì tiếp tục thực thi, ngược lại sẽ ném ra một AssertionError.


````python
x = 5
assert x > 3, "Giá trị của x phải lớn hơn 0"
print("Giá trị của x là:", x)
````

##8. Function (Hàm)

 **What (Hàm là gì?)**
 - Hàm là một nhóm các câu lệnh có liên quan thực hiện một nhiệm vụ cụ thể.
 - Nó giúp phân chia chương trình thành các phần nhỏ hơn, dễ quản lý và hiểu.
 - Hàm giúp tổ chức mã nguồn một cách hiệu quả hơn và tạo ra mã có thể tái sử dụng.

 **Why (Tại sao sử dụng Hàm?)**
 - Phân chia chương trình thành các phần nhỏ giúp làm cho mã nguồn dễ đọc và dễ bảo trì.
 - Tổ chức code thành các hàm giúp tạo ra mã nguồn linh hoạt và tái sử dụng được.
 - Hàm cung cấp khả năng tái sử dụng mã nguồn, giảm thiểu việc lặp lại mã và tăng khả năng bảo trì.

**Định nghĩa (Definition)**
- Đầu tiên, cần định nghĩa một hàm bằng từ khóa `def`, theo sau là tên hàm và các tham số (nếu có), được đặt trong dấu ngoặc đơn.
- Sau đó là một dòng mô tả ngắn gọn về chức năng của hàm, gọi là "docstring".
- Tiếp theo là các câu lệnh trong hàm, mỗi câu lệnh được thụt vào một mức.

```python
def function_name(arguments):
    """Mô tả ngắn về chức năng của hàm"""
    statement1
    statement2
    ...
```

**Gọi hàm (Call)**
- Để sử dụng hàm, cần gọi bằng tên hàm kèm theo các đối số (nếu có), được đặt trong dấu ngoặc đơn.
- Các giá trị truyền cho các đối số được gọi là đối số (arguments).

```python
function_name("a value")
```

Trong ví dụ dưới đây, hàm `greet` nhận một tham số `name` và in ra một lời chào chứa tên đã cho. Bằng cách gọi `greet("Name")`, hàm sẽ in ra `"Xin chào, Name"`.

````python
#Ví dụ
def greet(name):
    """In ra lời chào với tên đã cho"""
    print("Xin chào,", name)
````
